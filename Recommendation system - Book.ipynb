```python
# convert data into collaberative matrix
df_books = df.pivot_table(index = 'User.ID',
                     columns = 'Book.Title',
                     values = 'Book.Rating').reset_index(drop = True)
```

```python
df_books
```

```python
df_User = df['User.ID'].unique()
df_User
```

```python
df_User_Id = pd.DataFrame(df_User,columns = ['User_Id'])
df_User_Id
```

```python
df_User_Id_1 = df_User_Id.sort_values('User_Id')
df_User_Id_1
```

```python
df_books.index = df_User_Id_1.User_Id.unique()
df_books
```

```python
# Fill NAN with 0
df_books.fillna(0, inplace = True)
df_books
```

```python
# Calculate cosine similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
books_sim = 1-pairwise_distances(df_books.values, metric = 'cosine')
books_sim
```

```python
books_sim_df = pd.DataFrame(books_sim)
books_sim_df
```

```python
# Set row and column index with User id
books_sim_df.index = df_User_Id_1.User_Id.unique()
books_sim_df.columns = df_User_Id_1.User_Id.unique()
books_sim_df
```

```python
np.fill_diagonal(books_sim,0)
books_sim_df
```

```python
# Most Similar records
books_sim_df.idxmax(axis=1)[20:30]
```
